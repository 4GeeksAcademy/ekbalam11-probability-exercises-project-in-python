## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np
import sympy as sp

def roll_dice():
    return random.randint(1,6)

def roll_two_dices():
    return roll_dice() + roll_dice()
```

```python
thousand_rolls_array=[]
even_results_and_higher_than_seven = []
i = 0
while i < 1000:
    roll_result = roll_two_dices()
    thousand_rolls_array.append(roll_result)
    if roll_result > 7 or roll_result % 2 == 0:
        even_results_and_higher_than_seven.append(roll_result)  
        print(f"in the roll number: {i}, the dices are: {even_results_and_higher_than_seven[-1]}")  
    i+=1

```

```python
probability = len(even_results_and_higher_than_seven)/1000
print(f"The probability of getting 7 or an even number when rolling two dices is: {probability}") 
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
```

```python
def pick_random_ball():
    random_pick = random.randint(1,59)
    return ball_box[random_pick]

print(pick_random_ball())
```

```python
ball_pick_experiment = []
for picks in range(1,1001):
    pick = pick_random_ball()
    if pick == "White":
        ball_pick_experiment.append("White")
    elif pick == "Red":
        ball_pick_experiment.append("Red")
    elif pick == "Green":
        ball_pick_experiment.append("Green")

print(ball_pick_experiment)
```

```python
white_picked_balls = ball_pick_experiment.count('White')
red_picked_balls = ball_pick_experiment.count('Red')
green_picked_balls = ball_pick_experiment.count('Green')
pick_white_probability = white_picked_balls/1000
pick_red_probability = red_picked_balls/1000
pick_green_probability = green_picked_balls/1000
print(f"picked balls: \n white: {white_picked_balls} probability: {pick_white_probability} \n red: {red_picked_balls} probability:{pick_red_probability}  \n green: {green_picked_balls} probability: {pick_green_probability}")
```

***Whats the probability of:***

*Take 3 white and 2 red*

```python
number_of_orders = 10 #possible combinations of 5 balls with 3 whites and 2 reds
number_favorable_cases = (pick_white_probability**3) * (pick_red_probability **2) * number_of_orders
number_posible_cases = 1
three_white_two_red_probability = number_favorable_cases/number_posible_cases
print(f"the probability of getting 3 white balls and 2 red balls is: {round(three_white_two_red_probability*100,2)} %")
```

*Take all of the same color*

```python
all_same_probability = (pick_white_probability**5 + pick_red_probability**5 + pick_green_probability**5)
print(f" the probability for all the 5 picked balls be the same color is: {round(all_same_probability*100,2)} %")
```
