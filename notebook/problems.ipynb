## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np
import sympy as sp

def roll_dice():
    return random.randint(1,6)

def roll_two_dices():
    return roll_dice() + roll_dice()
```

```python
roll_two_dices()
```

```python
thousand_rolls_array=[]
even_results_and_higher_than_seven = []
i = 0
while i < 1000:
    roll_result = roll_two_dices()
    thousand_rolls_array.append(roll_result)
    if roll_result > 7 or roll_result % 2 == 0:
        even_results_and_higher_than_seven.append(roll_result)  
        print(f"in the roll number: {i}, the dices are: {even_results_and_higher_than_seven[-1]}")  
    i+=1

```

```python
probability = len(even_results_and_higher_than_seven)/1000
print(f"The probability of getting 7 or an even number when rolling two dices is: {probability}") 
```
